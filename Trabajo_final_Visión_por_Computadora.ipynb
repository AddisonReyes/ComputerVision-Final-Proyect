```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 as cv
```

```python
img = cv.imread("STScI-01GYZ9WMKF48Q091A5N02G7JKQ.tif")
img_fixed = cv.resize(img, (1200, 640))
```

```python
cv.imshow("Galaxia", img_fixed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(cv.cvtColor(img_fixed, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

# Identificar la cantidad de objetos

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
_, threshold_img = cv.threshold(gray_img, 200, 255, cv.THRESH_BINARY)

contours, _ = cv.findContours(threshold_img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
num_objects = len(contours)

plt.figure(figsize=(10, 6))
plt.imshow(cv.cvtColor(img_fixed, cv.COLOR_BGR2RGB))
plt.text(10, 30, f'Cantidad de objetos: {num_objects}', color='white', fontsize=12,
         bbox=dict(facecolor='black', alpha=0.7))
plt.axis('off')
plt.show()
```

# Realizar la separaci√≥n de los colores (Rojo, Verde y Azul)

```python
b, g, r = cv.split(img)
```

```python
b_fixed = cv.resize(b, (1200, 640))
cv.imshow('Canal Azul', b_fixed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
g_fixed = cv.resize(g, (1200, 640))
cv.imshow('Canal Verde', g_fixed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
r_fixed = cv.resize(r, (1200, 640))
cv.imshow('Canal Rojo', r_fixed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(131), plt.imshow(b, cmap='Blues'), plt.title('Canal Azul')
plt.axis('off')

plt.subplot(132), plt.imshow(g, cmap='Greens'), plt.title('Canal Verde')
plt.axis('off')

plt.subplot(133), plt.imshow(r, cmap='Reds'), plt.title('Canal Rojo')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Realizar la combinaci√≥n de los colores

```python
merged_img = cv.merge((b, g, r))
```

```python
merged_img_fixed = cv.resize(merged_img, (1200, 640))
cv.imshow('Canal Azul', merged_img_fixed)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(cv.cvtColor(img_fixed, cv.COLOR_BGR2RGB)), plt.title('Imagen Original')
plt.axis('off')

plt.subplot(122), plt.imshow(cv.cvtColor(merged_img, cv.COLOR_BGR2RGB)), plt.title('Imagen Combinada')
plt.axis('off')

plt.tight_layout()
plt.show()
```
